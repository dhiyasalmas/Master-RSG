```python
#Call Libary
import pandas as pd
import glob
import os
from pathlib import Path
import numpy as np
from google.colab import data_table
data_table.enable_dataframe_formatter()

#Insert data
from google.colab import drive
drive.mount("/content/gdrive")
# Read file txt from directory
path = r'/content/gdrive/My Drive/......' # use your path
files = Path(path).glob('*.txt')  # .rglob to get subdirectories
count_f = glob.glob(path + "/*.txt")
#count file thats can read
#print(len(count_f))

# made all file become 1 data frames
dfs = list()
for x in files:
  for i in range (len(count_f)):
    data = pd.read_csv(x, error_bad_lines=False)
    #print(data)
    # .stem is method for pathlib objects to get the filename w/o the extension
    data['file'] = x.stem
    dfs.append(data)
data_origen = pd.concat(dfs, ignore_index=True)

#Split, clean, store data
data_baru = []
def clean_data(data_masuk):
  #data_baru = []
  file_name = str(input("File name  without extension : "))
  for i in range (len(data_masuk)):
    kolom = data_masuk.iloc[i]
    i = i + 1
    #print(kolom)
    for j in range (len(kolom)):
      baris =  kolom.iloc[j]
      if baris != file_name:
        #if baris != 'Hargaawal':
          #if baris != file_name:
        split_data = np.char.split(baris," ").tolist()
        np_arr = np.array(split_data)
        index_must_deleted = np.argwhere(np_arr=='')
        fix_data = np.delete(np_arr, index_must_deleted)
        #print(fix_data)
        data_baru.append(fix_data)
      else:
       break
    j = j + 1
clean_data(data_origen)

df = pd.DataFrame(data_baru)
df.head()
from google.colab import files
file_name = str(input("File name output with extension : "))
df.to_csv(file_name, encoding = 'utf-8-sig', index = False)
files.download(file_name)
```
